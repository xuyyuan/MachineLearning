```python
# 封装起来
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

def plot_learning_curve(algo, X_train, X_test, y_train, y_test): # 注意algo这个参数
    
    train_score = []
    test_score = []
    for i in range(1, len(X_train)+1):
        
        algo.fit(X_train[:i], y_train[:i])

        y_train_predict = algo.predict(X_train[:i])
        train_score.append(mean_squared_error(y_train[:i], y_train_predict))

        y_test_predict = algo.predict(X_test)
        test_score.append(mean_squared_error(y_test, y_test_predict))
    
    plt.plot([i for i in range(1, len(X_train)+1)], np.sqrt(train_score), label='train')
    plt.plot([i for i in range(1, len(X_train)+1)], np.sqrt(test_score), label='test')
    plt.legend()
    plt.axis([0, len(X_train)+1, 0, 4])
    plt.show()
    
plot_learning_curve(LinearRegression(), X_train, X_test, y_train, y_test)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures, StandardScaler

def PolynomialRegression(degree):
    return Pipeline([
        ('poly', PolynomialFeatures(degree=degree)),
        ('std_scaler', StandardScaler()),
        ('lin_reg', LinearRegression()),
    ])
```

```python
poly2_reg = PolynomialRegression(degree=2)
plot_learning_curve(poly2_reg, X_train, X_test, y_train, y_test)

# 误差减小了（稳定在1附近）
```

```python
poly20_reg = PolynomialRegression(degree=20)
plot_learning_curve(poly20_reg, X_train, X_test, y_train, y_test)

# 测试数据集的误差与训练数据集之间的差距较大，通常是过拟合的情况（在训练数据集上表现得很好，在测试数据集上表现却不好， 模型的泛化能力较差）
```

### 测试train_test_split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=666)
```

```python
from sklearn.neighbors import KNeighborsClassifier

best_score, best_k, best_p = 0, 0, 0
for k in range(1, 11):
    for p in range(1, 6):
        knn_clf = KNeighborsClassifier(weights='distance', n_neighbors=k, p=p)
        knn_clf.fit(X_train, y_train)
        score = knn_clf.score(X_test, y_test)
        if score > best_score:
            best_score, best_k, best_p = score, k, p
print('Best K  = %s' % best_k )
print('Best P = %s' % best_p)
print('Best Score = %s'% best_score)
```

```python
# 以上使用train_test_split的方式来进行超参数的调整，下面使用交叉验证的方式进行调整
```

### 使用交叉验证

```python
# 不从底层实现了，直接从sklearn中实现
# 回头自己实现一遍吧，在如何分组上有一些复杂的技巧（可以先使用完全随机的方式进行尝试）
from sklearn.model_selection import cross_val_score

knn_clf = KNeighborsClassifier()
cross_val_score(knn_clf, X_train, y_train)
```

```python
# 准备正式实现下
best_score, best_k, best_p = 0, 0, 0
for k in range(2, 11):
    for p in range(1, 6):
        knn_clf = KNeighborsClassifier(weights='distance', n_neighbors=k, p=p)
        scores = cross_val_score(knn_clf, X_train, y_train)
        score = np.mean(scores)
        if score > best_score:
            best_score, best_k, best_p = score, k, p

print('Best_k = ', best_k)
print('Best_p = ', best_p)
print('Best_score = ', best_score)

# 使用交叉验证的方式获得的best_score要比train_test_split得到的best_score要小，但我们会更相信通过交叉验证方式得到的。
# 接上，因为在train_test_split中得到的参数很可能是过拟合了X_test(测试数据集)

# 我们这个交叉验证的目的是为了获得best_k和best_p，接下来就可以求得准确率
```

```python
best_knn_clf = KNeighborsClassifier(weights='distance', n_neighbors=k, p=p)
```

```python
best_knn_clf.fit(X_train, y_train)
best_knn_clf.score(X_test, y_test)

# 用三交叉验证的方式找到了最佳的参数组合，此时我们的模型准确度是0.980528511821975（因为是用一组完全未知的数据进行测量的，所以准确度是值得相信的）
```

## 偏差方差权衡 (Vias ViranceTrade off)

```python
# 偏差与方差是与我们的模型相关的

# 导致偏差的原因：对问题的假设不正确！如：非线性数据使用线性回归（欠拟和underfitting）；训练数据使用的特征和这个问题完全没有关系（如考试成绩与姓名）

# 数据一点点的扰动都会较大地影响模型。通常原因，使用地模型太复杂，如：高阶多项式（过拟合overfitting）

# 有一些算法天生是高方差的算法。如：kNN。kNN对数据比较敏感，因为它每次都去找与我们预测的样本最近的几个样本，让他们进行投票，但是离他最近的几个样本一旦有多数是错误的结果，那么我们预测的结果就不准确啦

# 通常非参数学习的算法都是高方差的算法，因为其不对数据进行任何假设，只能根据我们现有的训练数据来进行相应的预测，所以其对训练数据的准确性依赖程度较高（敏感），如kNN高度依赖样本数据，而并不完全依赖某些特定的参数

# 有些算法天生是高偏差的，如线性回归。现实生活中，很多问题本身就不是线性的，如果一定使用线性的，那么结果就会有偏差错误。

# 通常的参数学习都是高偏差算法，因为堆数据具有极强的假设。

# 那么问题就来了，大多数算法都具有参数，可以调整偏差和方差，如kNN中的k，k越小，说明我们的模型越复杂，相应的模型的方差越大，偏差越小；k越大，说明我们的模型越简单，偏差越大，方差越小。
# 同样，如线性回归中使用多项式回归，可以调整阶数degree，相应地2就会改变方差偏差，degree越小，模型越简单，相应地偏差就越大，degre越大，拟合地曲线越弯曲。

# 偏差和方差通常是互相矛盾的，降低偏差会提高方差，降低方差会提高偏差。理想情况下，低方差和低方差是最好的，可实际情况下，不怎么理想。

# 机器学习领域，主要的挑战是方差！但局限于算法这个层面。（课程中，假设有比较好的数据，数据也有相对比较好的特征），我们可以使我们的模型非常复杂，而让偏差非常低，但是方差大，泛化能力差，最终没有好的性能

# 解决方差的主要手段：1.降低模型复杂度；2.减少数据维度，降噪；3.增加样本数；4.使用验证集；5.模型的正则化

# 下面要讲的逻辑回归、SVM算法是离不开模型正则化的
```

## LASSO Regression （Regularization）

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)
y = 0.5 * x + 3 + np.random.normal(0, 1, size=100)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=666)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline

lin_reg = LinearRegression()
def PolynomialRegression(degree):
    return Pipeline([
        ('poly', PolynomialFeatures(degree=degree)),
        ('std_scaler', StandardScaler()),
        ('lin_reg', lin_reg)
    ])
```

```python
from sklearn.metrics import mean_squared_error

poly_reg = PolynomialRegression(degree=20)
poly_reg.fit(X_train, y_train)

y_poly_predict = poly_reg.predict(X_test)
mean_squared_error(y_test, y_poly_predict)
```

```python
def plot_model(model):
    X_plot = np.linspace(-3, 3, 100).reshape(100, 1)
    y_plot = model.predict(X_plot)
    plt.scatter(x, y)
    plt.plot(X_plot[:, 0], y_plot, color='r')
    plt.axis([-3, 3, 0, 6])
    plt.show()
    
plot_model(poly_reg)
```

### LASSO

```python
from sklearn.linear_model import Lasso

def LassoRegression(degree, alpha):
    return Pipeline([
        ('poly', PolynomialFeatures(degree=degree)),
        ('std_scaler', StandardScaler()),
        ('lasso_reg', Lasso(alpha=alpha))
    ])
```

```python
lasso_reg = LassoRegression(20, 0.01) # 这里的alpha要比岭回归要小一点，一般可以根据经验来写
lasso_reg.fit(X_train, y_train)

y1_predict = lasso_reg.predict(X_test)
mean_squared_error(y_test, y1_predict)
```

```python
plot_model(lasso_reg)
```

```python
lasso2_reg = LassoRegression(20, 0.1) 
lasso2_reg.fit(X_train, y_train)

y2_predict = lasso2_reg.predict(X_test)
mean_squared_error(y_test, y2_predict)

# 看误差减小了
```

```python
plot_model(lasso2_reg)
```

```python
lasso3_reg = LassoRegression(20, 1) 
lasso3_reg.fit(X_train, y_train)

y3_predict = lasso3_reg.predict(X_test)
mean_squared_error(y_test, y3_predict)

# 均方误差变大了
```

```python
plot_model(lasso3_reg)

# 此时，正则化的程度已经很高了（正则化过头）
# 实际选择这个机器学习参数的时候，是要在完全不正则化与正则化过头结果之间选择正则程度较好的这样的一个结果
```

### 弹性网 Elastic Net

```python
# 总结下岭回归与lasso回归
# 岭回归的计算是相对精准的，但是如果特征比较多， 整体的计算量会比较大；此时优先使用弹性网
# 弹性网结合了岭回归与lasso回归。lasso回归的缺点是急于将某些theta化为零，这个过程可能会出现一些错误，使得模型偏差较大
```
